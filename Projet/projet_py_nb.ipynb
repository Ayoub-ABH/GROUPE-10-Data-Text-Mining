Suppression des ponctuations, des chiffres et des caractÃ¨res spÃ©ciaux.

```python
GB_comments['comment_text'] = GB_comments['comment_text'].str.replace("[^a-zA-Z#]", " ")
print(GB_comments)
```

Suppression de mots courts.

```python
GB_comments['comment_text'] = GB_comments['comment_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
print(GB_comments)
```

Changer le texte en minuscules.

```python
GB_comments['comment_text'] = GB_comments['comment_text'].apply(lambda x:x.lower())
print(GB_comments)
```

TokÃ©nisation

```python
tokenized_tweet = GB_comments['comment_text'].apply(lambda x: x.split())
tokenized_tweet.head()
```

Lemmatisation

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
wnl = WordNetLemmatizer()
```

```python
tokenized_tweet.apply(lambda x: [wnl.lemmatize(i) for i in x if i not in set(stopwords.words('english'))]) 
tokenized_tweet.head()
```

```python
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
```

```python
GB_comments['comment_text'] = tokenized_tweet
print(GB_comments)
```

Faisons l'analyse des sentiments sur l'ensemble de donnÃ©es

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

 DÃ©finir les scores de sentiment

```python
GB_comments['Sentiment Scores'] = GB_comments['comment_text'].apply(lambda x:sia.polarity_scores(x)['compound'])
print(GB_comments)
```

```python
GB_comments.head()
```

Classer les scores de Sentiment comme Positif, NÃ©gatif et Neutre

```python
GB_comments['Sentiment'] = GB_comments['Sentiment Scores'].apply(lambda s : 'Positive' if s > 0 else ('Neutral' if s == 0 else 'Negative'))
print(GB_comments)
```

```python
GB_comments.head()
```

```python
GB_comments.Sentiment.value_counts()
```

Nous allons maintenant calculer le pourcentage de commentaires positifs de chaque vidÃ©o.

```python
videos = []
for i in range(0,GB_comments.video_id.nunique()):
    a = GB_comments[(GB_comments.video_id == GB_comments.video_id.unique()[i]) & (GB_comments.Sentiment == 'Positive')].count()[0]
    b = GB_comments[GB_comments.video_id == GB_comments.video_id.unique()[i]]['Sentiment'].value_counts().sum()
    Percentage = (a/b)*100
    videos.append(round(Percentage,2))
print(videos)
```

Nous allons maintenant calculer le pourcentage de commentaires nÃ©gatifs de chaque vidÃ©o.

```python
videos = []
for i in range(0,GB_comments.video_id.nunique()):
    a = GB_comments[(GB_comments.video_id == GB_comments.video_id.unique()[i]) & (GB_comments.Sentiment == 'Negative')].count()[0]
    b = GB_comments[GB_comments.video_id == GB_comments.video_id.unique()[i]]['Sentiment'].value_counts().sum()
    Percentage = (a/b)*100
    videos.append(round(Percentage,2))
print(videos)
```

Nous allons maintenant calculer le pourcentage de commentaires neutres de chaque vidÃ©o.

```python
videos = []
for i in range(0,GB_comments.video_id.nunique()):
    a = GB_comments[(GB_comments.video_id == GB_comments.video_id.unique()[i]) & (GB_comments.Sentiment == 'Neutral')].count()[0]
    b = GB_comments[GB_comments.video_id == GB_comments.video_id.unique()[i]]['Sentiment'].value_counts().sum()
    Percentage = (a/b)*100
    videos.append(round(Percentage,2))
print(videos)
```

Faire une dataframe des vidÃ©os avec leurs pourcentages positifs.

```python
Positivity = pd.DataFrame(videos,GB_comments.video_id.unique()).reset_index()
```

```python
Positivity.columns = ['video_id','Positive Percentage']
```
