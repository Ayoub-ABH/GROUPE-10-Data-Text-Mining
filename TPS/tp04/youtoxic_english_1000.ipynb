# Removal stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python

stop_words = stopwords.words('english')
```

# Tokenization

```python
def toknization(text):
    return text.split(" ")
```

```python
df['Text']=df['Text'].apply(toknization)
```

```python
df
```

# Removal stopwords

```python
def removestopwords(text):
    temp=[]
    for i in text:
        if i not in stop_words:
            temp.append(i)
    return temp
```

```python
df['Text']=df['Text'].apply(removestopwords)
```

```python
df
```

# stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemminnng(text):
    temp=[]
    for w in text:
       temp.append(ps.stem(w))
    return temp
```

```python
df['Text']=df['Text'].apply(stemminnng)
df
```
